```python
num1 = int(input("Enter the first number:"))
num2 = int(input("Enter the second number:"))
while True:
    ch = int(input('''
    Enter the operation number to perform
    1. Addition
    2. Multiplication
    3. Subtracion
    4. Division
    5. Exit
    '''))
    match ch:
        case 1:
            print("\n",num1 + num2)
        case 2:
            print("\n",num1 * num2)
        case 3:
            print("\n",num1 - num2)
        case 4:
            if num2 == 0:
                print("Invalid value")
            else:
                print(f"reminder {num1%num2} quotient {(num1/num2):.2f}")
        case 5:
            print("Thank you for using")
            break
        case _:
            print("Invalid choice")








```
