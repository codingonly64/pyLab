```python
# Define a function called sum_two_numbers(a, b) that takes two numbers as input and returns
# their sum. The program should take user input for two numbers and call the function to display
# the sum.

def sum_two_numbers(a, b):
    return a+b

num1 = float(input("Enter the first number:"))
num2 = float(input("Enter the second number:"))
result = sum_two_numbers(num1,num2)
print("The sum of the two numbers is", result)
```
