```python
# Define a function called factorial(n) that takes a number n as input and returns its factorial.
# The program should take user input for a number and call the function to display the factorial.


def factorial(n):
    return 1 if n <= 1 else  n * factorial(n - 1)

num = int(input("Enter a number: "))

print(f"The factorial of {num} is: {factorial(num)}")
```
