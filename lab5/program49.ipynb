```python
# Define a function called fibonacci(n) that returns the first n numbers in the Fibonacci sequence.
# The program should take user input for n and call the function to display the sequence.

def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence[:n]  

n = int(input("Enter the number: "))

print(f"The first {n} Fibonacci numbers are: {fibonacci(n)}")
```
