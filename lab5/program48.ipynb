```python
# Define a function swap(x, y) that takes two variables x and y, swaps their values, and then returns
# the updated values of x and y. It then swaps the values of two variables a and b using the swap
# function and prints the result.\

def swap(x, y):
    return y, x

a = int(input("Enter the first number (a): "))
b = int(input("Enter the second number (b): "))

a, b = swap(a, b)

print(f"After swapping: a = {a}, b = {b}")
```
