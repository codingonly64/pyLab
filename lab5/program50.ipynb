```python
# Define a function called area_of_circle(radius) that calculates the area of a circle given its radius.
# The program should take user input for the radius and call the function to display the area.

def area_of_circle(radius):
    return 3.14 * radius ** 2

radius = float(input("Enter the radius of the circle: "))

area = area_of_circle(radius)
print(f"The area of the circle with radius {radius} is: {area:.2f}")
```
