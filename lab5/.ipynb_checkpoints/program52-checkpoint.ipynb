```python
# Define a function called sum_of_digits(number) that takes an integer as input and returns the sum
# of its digits. The program should take user input for a number and call the function to display the
# sum of its digits.


def sum_of_digits(number):
    return sum(int(digit) for digit in str(abs(number)))
number = int(input("Enter an integer: "))

print(f"The sum of the digits of {number} is: {sum_of_digits(number)}")
```
