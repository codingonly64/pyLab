```python
# Define a function called isPrime(number) to check if a given number is prime. The program
# should take user input for a number and calls the isPrime function to check if that number is
# prime.

def isPrime(number):
    if number == 1:
        return False
    for i in range(2, int(number ** 0.5) ):
        if number % i == 0:
            return False
    return True

num = int(input("Enter the number:"))
result = isPrime(num)
print(num, "is Prime" if result else "is not Prime")
```
