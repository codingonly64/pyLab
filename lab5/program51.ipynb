```python
# Define a function called is_even(number) that returns True if the number is even and False if the
# number is odd. 
# The program should take user input for a number and call the function to display whether it's even or odd.

def is_even(number):
    return number % 2 == 0

number = int(input("Enter a number: "))

print(number, "is even." if is_even(number) else "is odd.")
```
