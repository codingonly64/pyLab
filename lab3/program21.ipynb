```python
def add(num1, num2):
	print(f"{num1} + {num2} = {num1 + num2}")
	menu()

def subtract(num1, num2):
	print(f"{num1} - {num2} = {num1 - num2}")
	menu()
	
def multiply(num1, num2):
	print(f"{num1} * {num2} = {num1 * num2}")
	menu()
	
def divide(num1, num2):
	print(f"{num1} / {num2} is  quotient {num1 / num2} and reminder is {num1 % num2}")
	menu()
	
def menu():
	num1 = int(input("Enter the number1: "))
	num2 = int(input("Enter the number2: "))
	print('''
			MENU
	1. add
	2. subtract
	3. multiply
	4. divide
	5. exit
	''')
	choice = int(input("Enter the choice"))
	match choice:
		case 1:
			add(num1, num2)
		case 2:
			subtract(num1, num2)
		case 3:
			multiply(num1, num2)
		case 4:
			divide(num1, num2)
		case 5:
			print("Thank you for using!!")
			return
		case _:
			print("Invlaid request!!")
			menu()
menu()
```
