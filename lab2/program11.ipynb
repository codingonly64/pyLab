```python
# Create a program that takes the radius of a circle as input and calculates the area and
# circumference
rad :float = float(input("Enter the radius of circle: "))
area = 3.14*rad*rad
circumference = 2*3.14*rad
print(f"Area of circle of radius {rad} is {area}")
print(f"Circumference of circle of radius {rad} is {circumference}")
```
