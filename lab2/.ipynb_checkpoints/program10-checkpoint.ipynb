```python
# Write a program to compute the average of three numbers provided by the user
num1 = int(input("Enter the first number:"))
num2 = int(input("Enter the second number:"))
num3 = int(input("Enter the third number:"))
avg = (num1+num2+num3)/3
print(f"Average of the numbers {num1}, {num2}, {num3} is {avg}")
```
