```python
# Write a program that takes a userâ€™s birth year and calculates their current age
from datetime import datetime
current_year = datetime.now().year
while True:
    birth_year = int(input("Enter the year: "))
    if birth_year >1900 and birth_year < current_year:
        break
    else:
        print("Invalid input")
current_age = current_year - birth_year
print(f"Your age is {current_age}")
```
