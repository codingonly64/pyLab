```python
# Write a program to check if a given number is an Armstrong number.
num = int(input("Enter the number:"))
temp = num
sm = 0
po = len(str(num))
while temp > 0:
    l_num = temp % 10
    sm += l_num ** po
    temp //= 10
if sm == num:
    print(num, "is an armstrong number")
else:
    print(num, "It is not an armstrong number")
```
