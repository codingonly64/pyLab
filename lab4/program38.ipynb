```python
# Print the prime numbers from 1 to n (nested loop)
num = int(input("Enter the number: "))
print("The prime numbers for range (1,",num,") is ", end="")
for i in range(2, num+1):
    is_prime = True
    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i, end=" ")
```
