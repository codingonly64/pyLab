```python
# Write a program that accepts a string and counts the number of vowels and consonants in it.
string = input("Enter the string:")
vowels = "aeiou"
v_count = 0
c_count = 0
temp = string.lower()
for i in temp:
    if i.isalpha():
        if i in vowels:
            v_count += 1
        else:
            c_count += 1
print("The number of vowels is", v_count, "and number of consoants is", c_count)
```
