```python
# Write a program to check whether a given number is a palindrome or not.
orginal_num = int(input("Enter the number:"))
reverse_num = 0
temp = orginal_num
while temp > 0:
    l_num = temp%10
    reverse_num = reverse_num * 10 + l_num
    temp //= 10
if orginal_num == reverse_num:
    print(f"{orginal_num} is a palindrome")
else:
    print(f"{orginal_num} is not a palindrome")
```
